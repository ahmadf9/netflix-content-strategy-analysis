## 4. Data Cleaning

```python
# Isi missing values
df['director'] = df['director'].fillna('Unknown')
df['cast']     = df['cast'].fillna('Unknown')
df['country']  = df['country'].fillna('Unknown')
df['genres']   = df['genres'].fillna('Unknown')
df['language'] = df['language'].fillna('Unknown')

# Konversi date_added ke datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

# Ambil negara pertama saja
df['country_clean'] = df['country'].apply(lambda x: x.split(',')[0].strip())

# Budget & revenue hanya ada di movies, isi NaN dengan 0
df['budget']  = df['budget'].fillna(0)
df['revenue'] = df['revenue'].fillna(0)

print('Data setelah cleaning:', df.shape)
df[['title', 'type', 'year_added', 'genres', 'language', 'vote_average', 'popularity']].head()
```

### 5.2 Tren Kualitas Konten Netflix per Tahun (Rating & Popularitas)

```python
# Catatan: dataset ini di-sample 1000 konten per tahun secara merata,
# sehingga tren jumlah konten tidak informatif.
# Sebagai gantinya, kita analisis tren KUALITAS (rating & popularitas) per tahun.

yearly_quality = (
    df[df['vote_count'] > 10]
    .groupby('release_year')
    .agg(
        avg_rating=('vote_average', 'mean'),
        avg_popularity=('popularity', 'mean')
    )
    .reset_index()
)

fig, ax1 = plt.subplots(figsize=(12, 5))

# Rating (sumbu kiri)
ax1.plot(yearly_quality['release_year'], yearly_quality['avg_rating'],
         marker='o', color='#E50914', linewidth=2, label='Avg Rating')
ax1.set_xlabel('Tahun Rilis')
ax1.set_ylabel('Rata-rata Rating', color='#E50914')
ax1.tick_params(axis='y', labelcolor='#E50914')

# Popularitas (sumbu kanan)
ax2 = ax1.twinx()
ax2.plot(yearly_quality['release_year'], yearly_quality['avg_popularity'],
         marker='s', color='steelblue', linewidth=2, linestyle='--', label='Avg Popularity')
ax2.set_ylabel('Rata-rata Popularitas', color='steelblue')
ax2.tick_params(axis='y', labelcolor='steelblue')

# Gabungkan legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Tren Kualitas Konten Netflix per Tahun (Rating & Popularitas)',
          fontsize=13, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('02_tren_kualitas.png', dpi=150)
plt.show()
```

### 5.7 Kecepatan Akuisisi Konten (Release Year vs Year Added)

```python
# Hitung gap antara tahun rilis dan tahun masuk Netflix
acquisition = df[df['year_added'].notna()].copy()
acquisition['gap_tahun'] = acquisition['year_added'] - acquisition['release_year']

# Filter yang masuk akal (gap 0-30 tahun)
acquisition = acquisition[(acquisition['gap_tahun'] >= 0) & (acquisition['gap_tahun'] <= 30)]

print('Rata-rata gap (tahun rilis → masuk Netflix):')
print(acquisition.groupby('type')['gap_tahun'].mean().round(1))
```

```python
# Distribusi gap per tipe konten
plt.figure(figsize=(10, 5))
sns.histplot(data=acquisition, x='gap_tahun', hue='type',
             bins=30, palette=['#E50914', 'steelblue'], alpha=0.7)

plt.title('Distribusi Gap: Tahun Rilis → Masuk Netflix', fontsize=13, fontweight='bold')
plt.xlabel('Gap (Tahun)')
plt.ylabel('Jumlah Konten')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('07_akuisisi_konten.png', dpi=150)
plt.show()
```

```python
# Proporsi konten baru (gap <= 1 tahun) vs konten lama (gap > 1 tahun)
acquisition['kategori'] = acquisition['gap_tahun'].apply(
    lambda x: 'Konten Baru (≤1 tahun)' if x <= 1 else 'Konten Lama (>1 tahun)'
)

akuisisi_counts = acquisition['kategori'].value_counts()

plt.figure(figsize=(6, 6))
explode = [0.05] + [0] * (len(akuisisi_counts) - 1)
plt.pie(akuisisi_counts,
        labels=akuisisi_counts.index,
        autopct='%1.1f%%',
        colors=['#E50914', '#AAAAAA'],
        startangle=140,
        explode=explode)

plt.title('Proporsi Konten Baru vs Konten Lama di Netflix', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('08_baru_vs_lama.png', dpi=150)
plt.show()

print(akuisisi_counts)
```

### 5.9 Rating per Genre — Genre Mana yang Terbaik?

```python
# Explode genre lalu merge dengan vote_average
genre_rating = df[df['vote_count'] > 10][['genres', 'vote_average']].copy()
genre_rating = genre_rating.assign(genre=genre_rating['genres'].str.split(', ')).explode('genre')
genre_rating['genre'] = genre_rating['genre'].str.strip()
genre_rating = genre_rating[genre_rating['genre'] != 'Unknown']

# Rata-rata rating per genre + jumlah konten
genre_stats = (
    genre_rating.groupby('genre')
    .agg(
        avg_rating=('vote_average', 'mean'),
        jumlah_konten=('vote_average', 'count')
    )
    .reset_index()
)

# Filter genre yang punya minimal 10 konten supaya representatif
genre_stats = genre_stats[genre_stats['jumlah_konten'] >= 10]
genre_stats['avg_rating'] = genre_stats['avg_rating'].round(2)
genre_stats = genre_stats.sort_values('avg_rating', ascending=False)

print('Top 10 Genre dengan Rating Tertinggi:')
print(genre_stats.head(10).to_string(index=False))
```

```python
# Visualisasi Top 10 genre rating tertinggi
top_genre_rating = genre_stats.head(10)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_genre_rating['genre'][::-1],
                top_genre_rating['avg_rating'][::-1],
                color='#E50914', alpha=0.8)

# Label angka di ujung bar
for bar, val in zip(bars, top_genre_rating['avg_rating'][::-1]):
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2,
             f'{val:.2f}', va='center', fontsize=10)

plt.title('Top 10 Genre Berdasarkan Rating Penonton', fontsize=13, fontweight='bold')
plt.xlabel('Rata-rata Vote Average')
plt.ylabel('Genre')
plt.xlim(0, top_genre_rating['avg_rating'].max() + 0.5)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('10_rating_per_genre.png', dpi=150)
plt.show()
```

Insight:
1. DOMINASI MOVIE vs TV SHOW\
   Movie mendominasi library Netflix, namun TV Show mendapat\
   rating penonton lebih tinggi — menandakan kualitas konten\
   serial lebih diapresiasi oleh penonton.

2. TREN KUALITAS KONTEN\
   Rating rata-rata konten Netflix berfluktuasi tiap tahun.\
   Konten tahun-tahun tertentu memiliki popularitas jauh lebih\
   tinggi — mengindikasikan adanya tahun 'peak' produksi konten\
   berkualitas yang dapat dijadikan benchmark strategi ke depan.

3. GENRE GAP: Volume ≠ Kualitas \
   Drama & Comedy paling banyak diproduksi, namun genre niche\
   seperti Documentary dan History justru memiliki rating lebih\
   tinggi — Netflix over-produce genre populer tapi under-produce\
   genre berkualitas tinggi.

4. KECEPATAN AKUISISI KONTEN\
   Mayoritas konten masuk Netflix dalam 1-3 tahun setelah rilis.\
   Namun Netflix juga mengakuisisi konten lama (>5 tahun) untuk\
   memperkaya library — strategi hybrid antara konten baru & klasik.

5. KONTEN VIRAL vs BERKUALITAS\
   Ada kesenjangan antara konten yang populer dan yang berkualitas.\
   'Hidden Gem' — konten berkualitas tinggi tapi kurang dikenal —\
   berpotensi besar jika dipromosikan lebih aktif oleh Netflix.

REKOMENDASI:
1. Tingkatkan produksi TV Show original — format serial
   terbukti mendapat rating lebih tinggi dan mendorong
   subscriber untuk bertahan lebih lama (retention).

2. Percepat akuisisi konten Asia (Korea, Jepang,
   Indonesia) dalam 1 tahun pertama rilis untuk
   menangkap momentum hype di pasar lokal.
